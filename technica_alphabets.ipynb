```python
import cv2
import mediapipe as mp
import numpy as np
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
mp_pose = mp.solutions.pose
```

```python
cap = cv2.VideoCapture(0)
with mp_pose.Pose(
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as pose:
  while cap.isOpened():
    success, image = cap.read()
    

    # Flip the image horizontally for a later selfie-view display, and convert
    # the BGR image to RGB.
    image = cv2.cvtColor(cv2.flip(image, 1), cv2.COLOR_BGR2RGB)
    # To improve performance, optionally mark the image as not writeable to
    # pass by reference.
    image.flags.writeable = False
    results = pose.process(image)
    with mp_hands.Hands(
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as hands:
        result = hands.process(image)
    

    # Draw the pose annotation on the image.
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    mp_drawing.draw_landmarks(
        image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
    if result.multi_hand_landmarks:
     for hand_landmarks in result.multi_hand_landmarks:
        mp_drawing.draw_landmarks(
         image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
        list=[]
        for i,l in enumerate( hand_landmarks.landmark):
               image_height, image_width, _ = image.shape
               cx,cy=int(l.x*image_width),int(l.y*image_height)
               list.append([i,cx,cy])
        if list[7][2]>list[6][2] and list[11][2]>list[10][2]:
             if list[15][2]>list[14][2] and list[19][2]>list[18][2]:
                    if list[8][2]>list[4][2]:
                        cv2.putText(image,"A" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
                        
        if list[7][2]<list[6][2] and list[11][2]<list[10][2]:
             if list[15][2]<list[14][2] and list[19][2]<list[18][2]:
                    cv2.putText(image,"B" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
                    
                    
        if list[8][2]>list[7][2] and list[12][2]>list[11][2]:
             if list[16][2]>list[15][2] and list[20][2]>list[19][2]:
                     if list[8][2]<list[5][2] and list[12][2]<list[9][2]:
                         if list[16][2]<list[13][2] and list[20][2]<list[17][2]:        
                            cv2.putText(image,"C" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
                    
       
            
        if list[8][2]<list[7][2] and list[12][2]>list[11][2]:
             if list[16][2]>list[15][2] and list[20][2]>list[19][2]:
                     if list[8][2]<list[5][2] and list[12][2]<list[9][2]:
                         if list[16][2]<list[13][2] and list[20][2]<list[17][2]:
                            cv2.putText(image,"D" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
                            
        if list[8][2]>list[7][2] and list[12][2]>list[11][2]:
             if list[16][2]>list[15][2] and list[20][2]>list[19][2]:
                if list[4][2]>list[15][2]:              
                     cv2.putText(image,"E" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
                            
        if list[20][2]<list[17][2] and list[16][2]<list[13][2]:
             if list[12][2]<list[9][2] and list[8][2]>list[7][2]:                    
                    cv2.putText(image,"F" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
       
        if list[12][2]>list[9][2] and list[16][2]>list[13][2]:
             if list[20][2]>list[17][2] and list[8][2]<list[5][2]:
                    if list[4][2]>list[14][2]:
                          cv2.putText(image,"G" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
                            
        if list[16][2]>list[13][2] and list[20][2]>list[17][2]:
           if list[8][2]<list[5][2] and list[12][2]<list[9][2]:
              if list[11][1]-list[7][1]>10:
                     if list[12][1]-list[1][1]>60:
                       cv2.putText(image,"H" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
        if list[12][2]>list[9][2] and list[16][2]>list[13][2]:
             if list[20][2]<list[17][2] and list[8][2]>list[5][2]:
                    if list[4][2]<list[9][2]:            
                         cv2.putText(image,"I" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
        if list[16][2]>list[13][2] and list[20][2]>list[17][2]:
           if list[8][2]<list[5][2] and list[12][2]<list[9][2]:
               if list[12][1]-list[8][1]>=20:
                    if list[4][2]<list[15][2]:                    
                         cv2.putText(image,"K" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
        if list[8][2]<list[6][2] and list[11][2]>list[10][2]:
             if list[15][2]>list[14][2] and list[19][2]>list[18][2]:
                     if list[10][1]-list[4][1]>=50:
                             cv2.putText(image,"L" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
        if list[8][2]>list[5][2] and list[12][2]>list[9][2]:
              if list[16][2]>list[13][2]:
                    cv2.putText(image,"M" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
        if list[8][2]>list[5][2] and list[12][2]>list[9][2]:
              if list[4][2]>list[16][2]:            
                        cv2.putText(image,"N" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
        if list[16][2]>list[13][2] and list[20][2]>list[17][2]:
           if list[8][2]<list[5][2] and list[12][2]<list[9][2]:
              if list[11][1]-list[7][1]<10:
                     cv2.putText(image,"R" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
        if list[8][2]>list[7][2] and list[12][2]>list[11][2]:
             if list[16][2]>list[15][2] and list[20][2]>list[19][2]:
                    
                if list[4][2]<list[15][2]:                
                     cv2.putText(image,"S" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
        if list[16][2]>list[13][2] and list[20][2]>list[17][2]:
           if list[8][2]<list[5][2] and list[12][2]<list[9][2]:
              if list[11][1]-list[7][1]>10 and list[11][1]-list[7][1]<20 :
                     if list[12][1]-list[1][1]<60:
                            if list[8][2]<list[6][2] and list[11][2]>list[10][2]:
             if list[15][2]>list[14][2] and list[19][2]>list[18][2]:
                     if list[10][1]-list[4][1]<=10:    
                             cv2.putText(image,"T" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
                        
                            cv2.putText(image,"U" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
        if list[16][2]>list[13][2] and list[20][2]>list[17][2]:
           if list[8][2]<list[5][2] and list[12][2]<list[9][2]:
               if list[12][1]-list[8][1]>=20:
                    if list[4][2]>list[15][2]:                 
                        cv2.putText(image,"V" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
        if list[16][2]<list[13][2] and list[20][2]>list[17][2]:
           if list[8][2]<list[5][2] and list[12][2]<list[9][2]:
             cv2.putText(image,"W" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
        if list[8][2]>list[7][2] and list[11][2]>list[10][2]:
             if list[15][2]>list[14][2] and list[19][2]>list[18][2]:
                if list[5][2]>list[8][2]:
                            cv2.putText(image,"X" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
        
        if list[7][2]>list[6][2] and list[11][2]>list[10][2]:
             if list[15][2]>list[14][2] and list[17][2]>list[20][2]:
                    if list[17][1]-list[4][1]>=55:          
                         cv2.putText(image,"Y" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
         if list[8][2]<list[6][2] and list[11][2]>list[10][2]:
             if list[15][2]>list[14][2] and list[19][2]>list[18][2]:
                     if list[10][1]-list[4][1]<=10:    
                             cv2.putText(image,"Z" , ((200,15)),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0,255), 2)
                        
            
                       
                            
    cv2.imshow('MediaPipe Pose', image)    
   
    if cv2.waitKey(5) & 0xFF == 27:
      break
cap.release()
```
